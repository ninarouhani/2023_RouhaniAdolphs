```r
# load packages 
library(reshape2)
library(ggplot2)
library(gridExtra)
```

```r
# set working directory to dataFiles and read in files
setwd('../dataFiles')
subDate<-read.csv('autobioMem_topicModeling.csv')

# order date
subDate$datePlot <- factor(subDate$datePlot, c("JAN","FEB","MAR","APR","MAY","JUN","JUL","AUG","SEP","OCT","NOV","DEC"))

# subset memory collections
sd_2020 = subset(subDate, set=='2020') # 2020 memory Dec 2020
sd_2020_p1 = subset(subDate, set=='2020+1') # 2020 memory one year later
sd_2020_p2 = subset(subDate, set=='2020+2') # 2020 memory two years later
sd_2020_p2_r = subset(subDate, set=='2020+2_rep') # 2020 memory two years later, replication in new sample
sd_2021 = subset(subDate, set=='2021') # 2021 memory Dec 2021 
```

## topic modeling

```r
# 2020 memory (recalled December 2020)

# set image size 
options(repr.plot.width=15,repr.plot.height=12)

# arrange topics in order
sd_2020$topic <- factor(sd_2020$topic, c('covid','social','occupation','election','domestic','BLM','covid-onset'))

ggplot(sd_2020, aes(x = datePlot, y = entryCount, fill = topic)) + scale_y_continuous(limits = c(0,1020), expand = c(0, 0))+
    geom_bar(stat = "identity") + xlab('')+ylab('memory entry count')+scale_fill_manual(values=c("#ff3333","#ff9900","#3399ff","#ffcc00","#E6E6FA","#6600ff","#ff0066"))+
    theme(panel.background = element_rect(fill = "white"), panel.border = element_rect(color = "grey80",fill=NA),panel.grid.major.y = element_line(color = "grey80"),axis.line.x = element_line(colour = "grey80"),axis.text.x = element_text(color='black',size=27), axis.line.y = element_line(colour = "black"),axis.title = element_text(size=30),axis.text = element_text(size=30),axis.title.y = element_text(size=35),legend.text=element_text(size=30),legend.title=element_text(size=30),legend.position=c(0.89,0.85))
```
